```python
import nltk

nltk.download('treebank')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# Named Entity Recognition (NER)
sent = nltk.corpus.treebank.tagged_sents()[1]
```

```python
print(sent)
```

```python
print(nltk.ne_chunk(sent, binary=True))
```

```python
print(nltk.ne_chunk(sent, binary=False))
```
